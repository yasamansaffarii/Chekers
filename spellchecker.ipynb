```python
import transformers
from transformers import AutoConfig, AutoTokenizer, AutoModel, TFAutoModel
import csv
# v3.0
model_name_or_path = "HooshvareLab/bert-fa-zwnj-base"
config = AutoConfig.from_pretrained(model_name_or_path)
tokenizer = AutoTokenizer.from_pretrained(model_name_or_path)

model = TFAutoModel.from_pretrained(model_name_or_path) # For TF
#model = AutoModel.from_pretrained(model_name_or_path)


# opening the CSV file
with open('/content/data/faspell_main.csv', mode ='r') as file:
    
  # reading the CSV file
  csvFile = csv.DictReader(file)
  wer=0
  alldata=0
  # displaying the contents of the CSV file
  for lines in csvFile:
      print("****************************************************")
      alldata=alldata+1
      text=lines['#misspelt']
      print("misspelt=",text)
      output=tokenizer.tokenize(text)
      realoutput=tokenizer.tokenize(lines['corrected'])
      print("corrected=",output)
      print("realCorrected=", lines['corrected'])
      if realoutput != output:
        wer=wer+1
      print("WER=", wer,"/", alldata)
text = "ما در هوش‌واره معتقدیم با انتقال صحیح دانش و آگاهی، همه افرات میتواننت از ابزارهای هوشمند استفاده کنند. شعار ما هوش مصنوعی برای همه است."
output=tokenizer.tokenize(text)
print(output)
#['ما', 'در', 'هوش', '[ZWNJ]', 'واره', 'معتقدیم', 'با', 'انتقال', 'صحیح', 'دانش', 'و', 'آ', '##گاهی', '،', 'همه', 'افراد', 'میتوانند', 'از', 'ابزارهای', 'هوشمند', 'استفاده', 'کنند', '.', 'شعار', 'ما', 'هوش', 'مصنوعی', 'برای', 'همه', 'است', '.']
```
